```python
def TrainTestDataGenerator(train, test, switch):
    continuous_predictors = continuous_predictors_switch(switch)
    y_train = train["DX_bl"]
    X_train = train.drop(["DX_bl"], axis=1)
    y_test = test["DX_bl"]
    X_test = test.drop(["DX_bl"], axis=1)
    X_train[continuous_predictors] = (X_train[continuous_predictors] - np.mean(X_train)[continuous_predictors])/(np.std(X_train)[continuous_predictors])
    X_test[continuous_predictors] = (X_test[continuous_predictors] - np.mean(X_test)[continuous_predictors])/(np.std(X_test)[continuous_predictors])
    X_train = X_train.drop(['CDRSB_bl', 'MMSE_bl', 'RAVLT_learning_bl', 'RAVLT_forgetting_bl', 'RAVLT_perc_forgetting_bl', 'RAVLT_immediate_bl'], 1);
    X_test = X_test.drop(['CDRSB_bl', 'MMSE_bl', 'RAVLT_learning_bl', 'RAVLT_forgetting_bl', 'RAVLT_perc_forgetting_bl', 'RAVLT_immediate_bl'], 1);
    
    return X_train, y_train, X_test, y_test
```

## Functions for All Models and Fitting

```python
# Logistic Regression
def log_ovr(X_train, y_train, X_test, y_test):
    log_ovr = LogisticRegressionCV(Cs=10, cv = 5, multi_class='ovr')
    log_ovr.fit(X_train, y_train)
    ac_train = log_ovr.score (X_train,y_train)
    ac_test = log_ovr.score (X_test,y_test)
    return log_ovr, ac_train, ac_test

# CV Log Regression multinomil
def log_multi(X_train, y_train, X_test, y_test):
    log_multi = LogisticRegressionCV(Cs=10, cv = 5, multi_class = 'multinomial')
    log_multi.fit(X_train,y_train)
    ac_train = log_multi.score (X_train,y_train)
    ac_test = log_multi.score (X_test,y_test)
    return log_multi, ac_train, ac_test

# Linear discriminant analysis
def LDA(X_train, y_train, X_test, y_test):
    lda = LinearDiscriminantAnalysis()
    lda.fit(X_train,y_train)
    ac_train = lda.score (X_train,y_train)
    ac_test = lda.score (X_test,y_test)
    return lda, ac_train, ac_test

# Quadratic discriminant analysis
def QDA(X_train, y_train, X_test, y_test):
    qda = QuadraticDiscriminantAnalysis()
    qda.fit(X_train,y_train)
    ac_train = qda.score (X_train,y_train)
    ac_test = qda.score (X_test,y_test)
    return qda, ac_train, ac_test

# CV k-nearest neighbour
def kNN(X_train, y_train, X_test, y_test):
    knn = KNN()
    param = np.arange(1,20)
    param_grid = dict(n_neighbors = param)
    gs = GridSearchCV(KNN(), param_grid, cv=5, n_jobs=2)
    gs.fit(X_train,y_train)
    knn_opt = gs.best_estimator_
    ac_train = knn_opt.score (X_train,y_train)
    ac_test = knn_opt.score (X_test,y_test)
    return knn_opt, ac_train, ac_test

# CV tree classifier
def DT(X_train, y_train, X_test, y_test):
    n_estimators = np.arange(1, 20)
    param_grid = dict(max_depth = n_estimators)
    gs = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, n_jobs=2)
    gs.fit(X_train,y_train)
    tree_opt = gs.best_estimator_
    ac_train = tree_opt.score (X_train,y_train)
    ac_test = tree_opt.score (X_test,y_test)
    return tree_opt, ac_train, ac_test

# CV Random forest
def RF(X_train, y_train, X_test, y_test):
    max_features = np.arange(1,len(X_train.columns))
    max_depth = np.arange(1,20)                                 
    rf = RandomForestClassifier(n_estimators=50, max_depth=None)
    param_grid = {"n_estimators": max_features, "max_depth" : max_depth}
    gs = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1)
    gs.fit(X_train, y_train)
    rf_opt = gs.best_estimator_
    ac_train = rf_opt.score (X_train,y_train)
    ac_test = rf_opt.score (X_test,y_test)
    return rf_opt, ac_train, ac_test

# Adaboost
def boost(X_train, y_train, X_test, y_test):
    max_features = [2*x for x in range(1,8)]
    max_depth = np.arange(1,5)
    learning_rates = [0.05, 0.1, 0.5]

    param_grid = {'n_estimators':max_features, 'base_estimator__max_depth':max_depth, \
                  'learning_rate':learning_rates}
    DTC = DecisionTreeClassifier()
    ABC = AdaBoostClassifier(base_estimator = DTC)
    gs = GridSearchCV(ABC, param_grid, cv=5, n_jobs=-1)
    gs.fit(X_train, y_train)
    ABC_opt = gs.best_estimator_
    ac_train = ABC_opt.score (X_train,y_train)
    ac_test = ABC_opt.score (X_test,y_test)
    return ABC_opt, ac_train, ac_test

# PCA
def PCA_logistic(X_train, y_train, X_test, y_test):
    pca = PCA(n_components = X_train.shape[1])
    pca.fit(X_train)
    X_train_pca = pca.transform(X_train)
    var_ratio = pca.explained_variance_ratio_
    pca_dict = {}
    var_sum = 0
    pca_counter = 0
    for i in range(len(var_ratio)): 
        if var_sum < 0.9:
            pca_dict[('pc %d'%i)] = X_train_pca[:, i] 
            var_sum = var_sum + var_ratio[i] 
            pca_counter = pca_counter + 1
    dfpca = pd.DataFrame(data = pca_dict)

    X_train_pca = dfpca.values
    X_test_pca = pca.transform(X_test)
    X_test_pca = X_test_pca[:, 0:pca_counter]
    log_ovr_pca = LogisticRegressionCV(Cs=10, cv = 5, multi_class='ovr')
    log_ovr_pca.fit(X_train_pca, y_train)
        
    ac_train = log_ovr_pca.score (X_train_pca,y_train)
    ac_test = log_ovr_pca.score (X_test_pca,y_test)
    return log_ovr_pca, ac_train, ac_test
    
# Linear SVM
def LSVM(X_train, y_train, X_test, y_test):
    Cs = [10**x for x in range(-3,3)]
    svm = LinearSVC(penalty='l2')
    param_grid = dict(C = Cs)
    gs = GridSearchCV(svm, param_grid, cv=5, n_jobs=-1)
    gs.fit(X_train, y_train)
    svm_opt = gs.best_estimator_
    ac_train = svm_opt.score (X_train,y_train)
    ac_test = svm_opt.score (X_test,y_test)
    return svm_opt, ac_train, ac_test

# RBF
def RBFSVM(X_train, y_train, X_test, y_test):
    Cs = [10**x for x in range(-3,3)]
    gammas = np.linspace(0, 1, 21)
    tols = [10**x for x in range(-3,3)]
    param_grid = {'C': Cs, 'gamma': gammas, 'tol': tols}
    svm = SVC(kernel='rbf', decision_function_shape='ovr')
    gs = GridSearchCV(svm, param_grid, cv=5, n_jobs=-1)
    gs.fit(X_train, y_train)
    svm_opt = gs.best_estimator_
    ac_train = svm_opt.score (X_train,y_train)
    ac_test = svm_opt.score (X_test,y_test)
    return svm_opt, ac_train, ac_test

# Sigmoid SVM
def SIGSVM(X_train, y_train, X_test, y_test):
    Cs = [10**x for x in range(-3,3)]
    gammas = np.linspace(0, 1, 21)
    tols = [10**x for x in range(-3,3)]
    coef0s = np.linspace(0, 1, 11)
    param_grid = {'C': Cs, 'gamma': gammas, 'tol': tols, 'coef0': coef0s}
    svm = SVC(kernel='sigmoid', decision_function_shape='ovr')
    gs = GridSearchCV(svm, param_grid, cv=5, n_jobs=-1)
    gs.fit(X_train, y_train)
    svm_opt = gs.best_estimator_
    ac_train = svm_opt.score (X_train,y_train)
    ac_test = svm_opt.score (X_test,y_test)
    return svm_opt, ac_train, ac_test
# Elastic Nets
def EN(X_train, y_train, X_test, y_test):
    alpha = np.linspace(0, 10, 11)
    l1_ratios = [.1, .5, .7, .9, .95, .99, 1]
    en_opt = ElasticNetCV(l1_ratio = l1_ratios, alphas = alpha, fit_intercept=True)
    en_opt.fit(X_train, y_train)
    y_train_hat = np.round(en_opt.predict(X_train))
    y_test_hat = np.round(en_opt.predict(X_test))
    ac_train = accuracy_score(y_train, y_train_hat)
    ac_test = accuracy_score(y_test, y_test_hat)
    return en_opt, ac_train, ac_test

# Multi-layer Perceptron Classifier
def MLPC(X_train, y_train, X_test, y_test):
    mlpc = MLPClassifier(hidden_layer_sizes = 2, activation = 'logistic', learning_rate='adaptive')
    mlpc.fit(X_train, y_train)
    ac_train = mlpc.score (X_train,y_train)
    ac_test = mlpc.score (X_test,y_test)
    return mlpc, ac_train, ac_test

# Perceptron
def PCT(X_train, y_train, X_test, y_test):
    alphas = [10**x for x in range(-4,4)]
    pct = Perceptron(penalty='elasticnet', fit_intercept=True)
    param_grid = {'alpha': alphas}
    gs = GridSearchCV(pct, param_grid, cv=5, n_jobs=-1)
    gs.fit(X_train, y_train)
    pct_opt = gs.best_estimator_
    ac_train = pct_opt.score (X_train,y_train)
    ac_test = pct_opt.score (X_test,y_test)
    return pct_opt, ac_train, ac_test

# GaussianProcess
def GP(X_train, y_train, X_test, y_test):
    gp = GaussianProcessClassifier(multi_class='one_vs_rest', n_jobs=-1, max_iter_predict=2)
    gp.fit(X_train, y_train)
    ac_train = gp.score (X_train,y_train)
    ac_test = gp.score (X_test,y_test)
    return gp, ac_train, ac_test
```

```python
def run_test(data_description, X_train, y_train, X_test, y_test, models_function_names, models_names):
    train_score =[]
    test_score =[]
    meta_model = []
    zipped = zip(models_function_names, models_names)
    print("===================================================================")
    print("For models running on %s dataset"%data_description)
    print("===================================================================")
    for i,r in zipped:
        model, accuracy_train, accuracy_test = i(X_train, y_train, X_test, y_test)

        print('%s train accuracy is %0.3f, test accuracy is %0.3f'%(r, accuracy_train, accuracy_test))
        train_score.append(accuracy_train)
        test_score.append(accuracy_test)        
        if r != 'PCA with Logistic Regression' and accuracy_test > 0.5:
            meta_model.append(model)
    if len(meta_model) != 0:
        ''' Meta Model '''
        accuracy_train, accuracy_test = MM_Stacking(meta_model, X_train, y_train, X_test, y_test, 'M')
        train_score.append(accuracy_train)
        test_score.append(accuracy_test)
        accuracy_train, accuracy_test = MM_Stacking(meta_model, X_train, y_train, X_test, y_test, 'LR')
        train_score.append(accuracy_train)
        test_score.append(accuracy_test)
        Model_Name_All = model_name_all
    else:
        Model_Name_All = models_names

    x_pos = np.arange(len(Model_Name_All))
    width = 0.4
    fig, ax = plt.subplots(1, 1, figsize=(20, 5))
    ax.bar(x_pos, train_score, width, alpha=0.5, label="Train")
    ax.bar([p + width for p in x_pos], test_score, width, alpha=0.5, label="Test")
    ax.set_xticklabels(Model_Name_All, size=10)
    ax.set_xticks(x_pos+width/2)
    ax.set_ylabel('Accuracy')
    ax.set_title('Models Accuracy (%s)'%data_description)
    plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
    ax.legend(loc='best', fontsize = 15);
    plt.tight_layout()
    fig.savefig('Figures/%s'%data_description, format='eps', dpi=1000)
    return meta_model, train_score, test_score
```

## Meta Models

```python
def majority_voter(lst):
    return max(set(lst), key=lst.count)
def majority_voter_classifier(M): 
    result = []
    number_of_samples = M.shape[0]
    for i in range(number_of_samples):
        result.append(majority_voter(list(M[i, :])))
    return result
```

```python
def MM_Stacking(meta_model, X_train, y_train, X_test, y_test, flag):
    predictions_train = np.zeros((X_train.shape[0], len(meta_model))) 
    predictions_test = np.zeros((X_test.shape[0], len(meta_model)))

    for i in range(len(meta_model)):
        predictions_train[:, i] = np.round(meta_model[i].predict(X_train))
        predictions_test[:, i] = np.round(meta_model[i].predict(X_test))
        
    if flag == 'M': 
        predictions_train_mv = majority_voter_classifier(predictions_train)
        predictions_test_mv = majority_voter_classifier(predictions_test)
        ac_train_mv = accuracy_score(y_train, predictions_train_mv)
        ac_test_mv = accuracy_score(y_test, predictions_test_mv)

        print('=== Majority Vote as Meta-Model ===')
        print('train accuracy: %f' %ac_train_mv) 
        print('test accuracy: %f' %ac_test_mv)
        return ac_train_mv, ac_test_mv
    
    if flag == 'LR':
        predictions_train_mv = majority_voter_classifier(np.round(predictions_train))
        predictions_test_mv = majority_voter_classifier(np.round(predictions_test))
        ac_train_lr = accuracy_score(y_train, predictions_train_mv)
        ac_test_lr = accuracy_score(y_test, predictions_test_mv)

        print('=== Majority Vote as Meta-Model ===')
        print('train accuracy: %f' %ac_train_lr) 
        print('test accuracy: %f' %ac_test_lr)
        return ac_train_lr, ac_test_lr
```

## Comparison of Datasets

```python
data_files = ['ADNI2 (Protein. ABETA, TAU, PTAU)-Model-Based Imputation.csv',
              'ADNI2 (Drug)-Model-Based Imputation.csv', 'ADNI2 (Drug and Protein)-Model-Based Imputation.csv']
path = 'Data/'
df_train = pd.read_csv(path+'ADNI2-Model-Based Imputation.csv')['Train'].values
df_test = pd.read_csv(path+'ADNI2-Model-Based Imputation.csv')['Test'].values
train_ac = [df_train[[1, 6, 10, 12, 14]]]
test_ac = [df_test[[1, 6, 10, 12, 14]]]
for file in data_files:
    path_file = path + file
    train_ac.append(pd.read_csv(path_file)['Train'].values[0:5])
    test_ac.append(pd.read_csv(path_file)['Test'].values[0:5])


file_description = ['ADNI2', 
#                     'ADNI2 (Protein: TAU)', 'ADNI2 (Protein: PTAU)',
#                     'ADNI2 (Protein: ABETA)', 
                    'ADNI2 (Protein: ABETA, TAU, PTAU)',
                    'ADNI2 (Drug)', 'ADNI2 (Drug and Protein)']
model_names = ['Logistic (Multinomial)', 'Random Forest', 'RBF SVM', 
               'Elastic Nets', 'Multi-layer Perceptron Classifier']

fig, ax = plt.subplots(1, 2, figsize=(26, 8))
x_pos = np.arange(len(file_description))
width = 0.2

figure_description = 'Comparison of Traing Accuracies over Datasets (Model-Based Imputation)'
for i in range(len(model_names)):
    model = model_names[i]
    ax[0].bar([p + i*width for p in x_pos], [ac[i] for ac in train_ac], width, alpha=0.5, label=model)
ax[0].set_xticklabels(file_description, size=10)
ax[0].set_xticks(x_pos+2*width)
ax[0].set_ylabel('Scores')
ax[0].set_title(figure_description)
plt.setp(ax[0].xaxis.get_majorticklabels(), rotation=0)
ax[0].set_ylim((0.6, 1))
plt.tight_layout()

figure_description = 'Comparison of Test Accuracies over Datasets (Model-Based Imputation)'
for i in range(len(model_names)):
    model = model_names[i]
    ax[1].bar([p + i*width for p in x_pos], [ac[i] for ac in test_ac], width, alpha=0.5, label=model)
ax[1].set_xticklabels(file_description, size=10)
ax[1].set_xticks(x_pos+2*width)
ax[1].set_ylabel('Scores')
ax[1].set_title(figure_description)
plt.setp(ax[1].xaxis.get_majorticklabels(), rotation=0)
ax[1].legend(loc='best', fontsize = 15);
ax[1].set_ylim((0.6, 1))
plt.tight_layout()

fig.savefig('Figures/%s'%'Comparison of Accuracies over Datasets (Model-Based Imputation)', format='eps', dpi=1000)

accuracies_dict = {}
for i in range(len(train_ac)):
    col1 = file_description[i] + '.Train'
    col2 = file_description[i] + '.Test'
    accuracies_dict[col1] = train_ac[i]
    accuracies_dict[col2] = test_ac[i]
ac = pd.DataFrame(accuracies_dict)
ac.to_csv(path_or_buf='Data/%s.csv'%'Comparison of Accuracies over Datasets (Model-Based Imputation)')
```
